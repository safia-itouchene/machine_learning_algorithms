```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder , OrdinalEncoder
```

```python
bc_data = load_breast_cancer()
```

```python
print(bc_data.keys())
```

```python
print(bc_data.DESCR)
```

```python
df_bc = pd.DataFrame(bc_data.data , columns=bc_data.feature_names)
df_bc.head()
```

```python
y = bc_data.target
y
```

```python
X_train , X_test , y_train ,y_test = train_test_split(df_bc , y , test_size=0.25 , stratify=y)
```

```python
#preprocessing
min_max_scaler = MinMaxScaler()
X_train_norm = min_max_scaler.fit_transform(X_train)
X_test_norm = min_max_scaler.fit_transform(X_test)
```

```python
# application of Logistic Regression
lr = LogisticRegression()
```

```python
# we give to the model data(input data) with labels(output data)
lr.fit(X_train_norm , y_train)
```

```python
""" 
lr.coef_ is a NumPy array that contains the weights assigned to each feature.
These weights determine how much each feature contributes to the model's 
prediction.
"""
lr.coef_
```

```python
""" 
lr.intercept_ is a bias value that the model adds to its prediction 
before considering the input features.
It helps the model make better predictions when all feature values are zero.
"""
lr.intercept_
```

```python
# train the model
y_pred = lr.predict(X_test_norm)
```

```python
print(accuracy_score(y_test,y_pred))
```
