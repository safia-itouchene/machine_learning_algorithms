```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist_data = load_digits()
```

```python
mnist_data.data.shape
```

```python
X_0 = mnist_data.data[0]
X_0_img = np.reshape(X_0,(8,8))
plt.imshow(X_0_img , cmap = "gray")
```

```python
X = mnist_data.data
X = X/X.max()
```

```python
centroids = [i for i in range(2,15)]
list_wcss = []

for centroid in centroids:
    k_mean = KMeans(centroid)
    k_mean.fit(X)
    wcss = k_mean.inertia_
    print(f"the within sum of squares for k ={centroid} is: {wcss}")
    list_wcss.append(wcss)
```

```python
#plotting Elbow method
plt.scatter(centroids , list_wcss)
```
