```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import OneHotEncoder ,OrdinalEncoder,LabelEncoder
from sklearn.multiclass import OneVsRestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from skmultilearn.problem_transform import BinaryRelevance
from skmultilearn.problem_transform import ClassifierChain
from skmultilearn.problem_transform import LabelPowerset
from sklearn.metrics import roc_auc_score
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
X = pd.read_csv('preprocessed_data.csv')
```

```python
X.head()
```

```python
y = pd.read_csv('data_target.csv')
```

```python
y.head()
```

```python
X_train ,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25,stratify=y)
```

```python
"""
If there are strong dependencies between the labels (e.g., patients with Diabete are more likely to also have Hypertension),
you can use ClassifierChain . This method chains the classifiers so that the prediction of one label influences the next.
"""
model =  ClassifierChain(XGBClassifier())
model.fit(X_train,y_train)
```

```python
y_pred = model.predict_proba(X_test)
```

```python
#  ROC_AUC_Score
score = roc_auc_score(y_test, y_pred.toarray(), average='macro')
print("Score = ",score)
print("\n")
```
