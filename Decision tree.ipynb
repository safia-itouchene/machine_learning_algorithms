```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplot
import pandas as pd
```

```python
iris_data = load_iris()
```

```python
iris_data.keys()
```

```python
"""
What is iris.feature_names?
It gives us the names of each column (feature).
What is iris.feature_names?
It gives us the names of each column (feature).
"""
df = pd.DataFrame(iris_data.data , columns= iris_data.feature_names)
```

```python
df.head()
```

```python
#put output data in y 
y = iris_data.target
```

```python
# stratify=y ensures that the proportion of classes in y remains the same in both train and test sets.
X_train,X_test,y_train,y_test = train_test_split(df,y,test_size=0.25,stratify=y)
```

```python
#tain the model
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
#test the model
y_pred = dt.predict(X_test)
```

```python
"""
accuracy_score(y_test, y_pred) :
Compares y_pred with y_test.
Calculates the percentage of correct predictions.
"""
print(accuracy_score(y_test,y_pred))
```

```python
plot_tree(dt , feature_names=iris_data.feature_names,class_names=iris_data.target_names,filled=True);
```
